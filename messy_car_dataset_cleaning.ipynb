# Task -2 Data Cleaning (Clean the Data Fix the issues identified)

```python
df['mileage'] = df['mileage'].str.strip().replace("kmpl","")
```

```python
df['mileage'] = df['mileage'].str.replace({"kmpl":""})
```

```python
df['mileage'] = pd.to_numeric(df["mileage"])

df['mileage'] = df['mileage'].fillna(df['mileage'].mean())

df.info()
```

dropped the Unamed : 0 column (No usefull)

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df
```

Removing the White spaces in the car name column

```python
df['car_name'] = df['car_name'].str.strip().str.title()

df['car_name']
```

Removing the White spaces in the brand column

```python

df['brand'] = df['brand'].str.strip().str.title()
df['brand'].unique()
```

Removing the White spaces in the Model column

```python
df['model'] = df['model'].str.strip().str.title()
df['model'].unique()
```

Cleaning the string values and commas to numeric values in Km_driven column

```python
print(df['km_driven'].sample(200).tolist())
```

```python
df[df['km_driven'].str.contains("km")]
```

```python
df['km_driven'] = df['km_driven'].str.strip().replace({"kms":"", "km":""})
```

```python
df['km_driven'] = pd.to_numeric(df['km_driven'].str.replace(',', ''), errors='coerce')
```

```python

df['km_driven'].dtype
```

```python
df['km_driven'] = df['km_driven'].fillna(df['km_driven'].mean())
```

```python
df.info()
```

```python
df.duplicated().sum()
```

Cleaning the seller_type column by changing the names

```python
df['seller_type'].unique()
```

```python
df['seller_type'] = df['seller_type'].str.replace({"Delaer":"dealer",
                                                    "Individul":"individual",
                                                    "Individuall":"individual"
                                                    }).str.title()
```

```python
df['seller_type'].unique()
# cleaned
```

Changing the cases and removing whitespaces

```python
df['fuel_type'] = df['fuel_type'].str.strip().str.title()
```

```python
df['fuel_type'].unique()
```

```python
df
```

```python
df.isnull().sum()
```

imputing the values in engine and max_power columns

```python
for col in ['engine', 'max_power']:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)
```

```python
df['seats'] = df['seats'].fillna(df['seats'].mode()[0])
```

```python
# df['selling_price'] = df['selling_price'].dropna()
df.info()
```

dropping the null values in the target variable

```python
df = df.dropna(subset=['selling_price'])

df.info()
```

```python
df.duplicated().sum()
df = df.drop_duplicates()

df.info()
```

```python
df.duplicated().sum()
```

# TASK - 3 Compute Baseline MAE

```python

import numpy as np

y = df['selling_price']

# Baseline Prediction

mean_price = y.mean()
y_pred = np.full(len(df) , mean_price)

# MAE Calculation
mae = np.mean(np.abs(y - y_pred))

print(mae)
```
